## Recommend post for given user

```python
post.head()
```

```python
post=post.rename(columns={'_id':'post_id'})
post
```

```python
user.head()
```

```python
user=user.rename(columns={'_id':'user_id'})
user
```

```python
view.head()
```

```python
view["Valuable"] = np.random.randint(1, 6, len(view))
```

```python
view.shape
```

```python
df1=pd.merge(view,user,on='user_id')
```

```python
df1.head()
```

```python
df=pd.merge(df1,post, on='post_id')
```

```python
df.head()
```

```python
col=['timestamp','gender','academics','category']
df=df.drop(col,axis=1)
df
```

```python
df_rating=df.dropna(axis=0, subset=['title'])
post_rating=(df_rating.
             groupby(by=['title'])['Valuable'].
             count().
             reset_index().
             rename(columns={'Valuable':'totalValuableCount'})
             [['title','totalValuableCount']]
            )
post_rating.head()
```

```python
rating=post_rating.merge(df_rating, left_on='title', right_on='title', how='left')
rating.head()
```

```python
rating.shape
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(post_rating['totalValuableCount'].describe())
```

```python
print(post_rating['totalValuableCount'].quantile(np.arange(.9,1,.01)))
```

```python
#we have to fix a threshold here that machine will recommend post to user that have votes greater than
# the threshold values
popularity_thresold=13
rating_majority=rating.query('totalValuableCount >= @popularity_thresold')
rating_majority.head()
```

```python
rating_majority.columns
```

```python
from scipy.sparse import csr_matrix
```

```python
rating_majority_post=rating_majority.drop_duplicates(['user_id',' post_type'])
rating_majority_post_pivot=rating_majority_post.pivot(index=' post_type',columns='user_id', values='Valuable').fillna(0)
rating_majority_post_matrix=csr_matrix(rating_majority_post_pivot.values)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model_knn=NearestNeighbors(metric='cosine', algorithm='brute')
model_knn.fit(rating_majority_post_matrix)
```

```python
rating_majority_post_pivot
```

## Choose a random users and recommend it a title of post by looking their cosine simliarity scores

```python
query_index = np.random.choice(rating_majority_post_pivot.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(rating_majority_post_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 2)
```
